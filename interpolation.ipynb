```
myimage = cv2.imread('icon.png')
cv2_imshow(myimage)
print(myimage.shape)
```

```
resize_img_interpolation = cv2.resize(myimage,(500,500),interpolation=cv2.INTER_LINEAR_EXACT)
cv2_imshow(resize_img_interpolation)
```

```
resize_img_interpolation = cv2.resize(myimage,(500,500),interpolation=cv2.INTER_LINEAR)
cv2_imshow(resize_img_interpolation)
```

```
resize_img_interpolation = cv2.resize(myimage,(500,500),interpolation=cv2.INTER_NEAREST)
cv2_imshow(resize_img_interpolation)
```

```
resize_img_interpolation = cv2.resize(myimage,(500,500),interpolation=cv2.INTER_AREA)
cv2_imshow(resize_img_interpolation)
```

```
resize_img_interpolation = cv2.resize(myimage,(500,500),interpolation=cv2.INTER_CUBIC)
cv2_imshow(resize_img_interpolation)
```

```
resize_img_interpolation = cv2.resize(myimage,(500,500),interpolation=cv2.INTER_LANCZOS4)
cv2_imshow(resize_img_interpolation)
```

```
resize_img_interpolation = cv2.resize(myimage,(500,500),interpolation=cv2.INTER_TAB_SIZE2)
cv2_imshow(resize_img_interpolation)
```

```
resize_img_interpolation = cv2.resize(myimage,(500,500),interpolation=cv2.INTER_MAX)
cv2_imshow(resize_img_interpolation)
```
